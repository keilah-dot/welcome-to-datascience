```python
import numpy as np
data_type = [('name', "S15"),('grade',int),('height',float)]
student_details=[('Pengui',6,145.0),('James',12,155.0),('Amanda',10,150.32),('Jeremy',9,155.65)]
student = np.array(student_details,dtype=data_type)
print(f"Original Array:\n{student}")
print(f"Sorted Array:\n{np.sort(student,order = 'height')}")
```
