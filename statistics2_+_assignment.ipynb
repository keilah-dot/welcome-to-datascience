```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Weather Dataset.csv")
df.head()
```

```python
df.info()
```

```python
mean_temp = np.mean (df["Temperature (C)"])
print(f"mean tempature is: {mean_temp}")
var_temp = np.var(df["Temperature (C)"])
print(f"Variance of Temp is: {var_temp}")
std_temp = np.std(df["Temperature (C)"])
print(f"Standard Deviation of Temp is: {std_temp}")
```

```python

for i in range(1, 13):
  month = df.loc[df["month"] == i]["Temperature (C)"]
  print("For month "+str(i))
  print("Mean temperature is "+ str(np.mean(month)))
  print("Standard deviation is "+ str(np.std(month))+"\n")
```
