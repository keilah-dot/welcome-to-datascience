```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Bestsellers with categories.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
mean_Userr = np.mean(df["User Rating"])
print(f"Mean User Rating: {mean_Userr}")
median_Userr = np.median(df["User Rating"])
print(f"Median User Rating: {median_Userr}")
mode_Userr = stats.mode(df["User Rating"])
print(f"Mode User Rating: {mode_Userr}")
```

```python
mean_Price = np.mean(df["Price"])
print(f"Mean future Price: {mean_Price}")
median_Price = np.median(df["Price"])
print(f"Median future Price: {median_Price}")
mode_Price = stats.mode(df["Price"])
print(f"Mode future Price: {mode_Price}")
```

```python
mean_Reviews = np.mean(df["Reviews"])
print(f"Mean future Reviews: {mean_Reviews}")
median_Reviews = np.median(df["Reviews"])
print(f"Median future Reviews: {median_Reviews}")
mode_Reviews = stats.mode(df["Reviews"])
print(f"Mode future Reviews: {mode_Reviews}")
```
